```python
import numpy as np
```

```python
def decode_segmap(image, nc=2):
    label_colors = np.array([(0, 0, 0), (128, 128, 128)])   # 0 - background, 1 - person
    r = np.zeros_like(image).astype(np.uint8)
    g = np.zeros_like(image).astype(np.uint8)
    b = np.zeros_like(image).astype(np.uint8)

    for l in range(0, nc):
      idx = image == 1
      r[idx] = label_colors[1, 0]
      g[idx] = label_colors[1, 1]
      b[idx] = label_colors[1, 2]

    rgb = np.stack([r, g, b], axis = 2)
    return rgb
```

```python
def segment(net, path):
    img = Image.open(path)
    plt.imshow(img)
    plt.axis('off')
    plt.show()
    trf = T.Compose([T.Resize(256), T.CenterCrop(224), T.ToTensor(), 
                 T.Normalize(mean = [0.485, 0.456, 0.406], std = [0.229, 0.224, 0.225])])
    inp = trf(img).unsqueeze(0)
    out = net(inp)['out']
    om = torch.argmax(out.squeeze(), dim=0).detach().cpu().numpy()
    rgb = rgb = decode_segmap(om)
    plt.imshow(rgb)
    plt.axis('off')
    plt.show()
```

```python
segment(fcn, img_path)
```
